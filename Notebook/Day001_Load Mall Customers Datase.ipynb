### Topic 3: Basic Summary & Insights

```python
# Get min, max, mean for each column
print("Age summary")
print("Min:",df['Age'].min())
print("Max :",df['Age'].max())
print("Mean is :",df['Age'].mean())
print("\nIncome Summary:")
print("Min:", df["Annual Income (k$)"].min())
print("Max:", df["Annual Income (k$)"].max())
print("Mean:", df["Annual Income (k$)"].mean())

print("\nSpending Score Summary:")
print("Min:", df["Spending Score (1-100)"].min())
print("Max:", df["Spending Score (1-100)"].max())
print("Mean:", df["Spending Score (1-100)"].mean())
```
